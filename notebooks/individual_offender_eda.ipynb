- 1 = White/Caucasian
- 2 = Black/African American

- 5 = Multi-racial
- 7 = Other

- 9 = Non-US American Indians
- 10 = American Indians Citizenship Unknown

### BOOKERCD Meaning:
- 0 = Within Range
- 1 = Upward Departure
- 2 = Upward Departure w/Booker
- 3 = Above Range w/Booker
- 4 = Remaining Above Range
- 5 = 5K1.1/Substantial Assistance
- 6 = Early Disposition/5K3.1
- 7 = Government Sponsored -Below Range
- 8 = Downward Departure
- 9 = Downward Departure w/Booker
- 10 = Below Range w/Booker
- 11 = Remaining Below Range
- ꞏ = Missing, Indeterminable, or Inapplicable

### SENTRNGE Meaning: 
- 0 = Within Range
- 1 = Upward Departure
- 2 = 5K1.1/Substantial Assistance
- 3 = Early Disposition/5K3.1
- 4 = Government Sponsored Departure
- 5 = Downward Departure
- 6 = Above Range Variance
- 7 = Government Sponsored Variance
- 8 = Below Range Variance
- ꞏ = Missing, Indeterminable, or Inapplicable 

### Conversions:
- BOOKER 0 = SENTRNGE 0
- BOOKER 1 = SENTRNGE 1
- BOOKER 2 = SENTRNGE 1
- BOOKER 3 = SENTRNGE 6
- BOOKER 4 = SENTRNGE 6
- BOOKER 5 = SENTRNGE 2
- BOOKER 6 = SENTRNGE 3
- BOOKER 7 = SENTRNGE 4
- BOOKER 8 = SENTRNGE 5
- BOOKER 9 = SENTRNGE 5
- BOOKER 10 = SENTRNGE 8 
- BOOKER 11 = SENTRNGE 8

Now that we have set up our conversions, let's update the SENTRNGE columns with their BOOKER equivalents.

```python
tn['SENTRNGE'] = tn['SENTRNGE'].fillna('Unknown')
tn.SENTRNGE.isnull().sum()
```

```python
sentrnge = []
for row_index, row_value in tn.iterrows(): 
    if row_value['SENTRNGE'] == 'See BOOKERCD':
        if row_value['BOOKERCD'] == 0:
            sentrnge += [0]
        elif str(row_value['BOOKERCD']) == '1':
            sentrnge += [1]
        elif str(row_value['BOOKERCD']) == '2':
            sentrnge += [1]
        elif str(row_value['BOOKERCD']) == '3':
            sentrnge += [6]
        elif str(row_value['BOOKERCD']) == '4':
            sentrnge += [6]
        elif str(row_value['BOOKERCD']) == '5':
            sentrnge += [2]
        elif str(row_value['BOOKERCD']) == '6':
            sentrnge += [3]
        elif str(row_value['BOOKERCD']) == '7':
            sentrnge += [4]
        elif str(row_value['BOOKERCD']) == '8':
            sentrnge += [5]
        elif str(row_value['BOOKERCD']) == '9':
            sentrnge += [5]
        elif str(row_value['BOOKERCD']) == '10':
            sentrnge += [8]
        elif str(row_value['BOOKERCD']) == '11':
            sentrnge += [8]
        else:
            sentrnge += ['Unknown']
    else:
        sentrnge.append(row_value['SENTRNGE'])
```

```python
tn['CRIMETYPE'] = tn['NWSTAT1'].astype(str).str[:2]
tn.value_counts('CRIMETYPE')
```

```python
crimtype = []
for row_index, row_value in tn.iterrows():
#     if row_value['ADJ_D1'] > 0:
#         crimtype += ['violent']
    if row_value['CRIMETYPE'] == '81':
        crimtype += ['immigration']
    elif row_value['NWSTAT1'][:5] == '18111':
        crimtype += ['murder']
    elif row_value['NWSTAT1'][:5] == '18120':
        crimtype += ['kidnapping']
    elif row_value['NWSTAT1'][:5] == '18922':
        crimtype += ['firearms']
    elif row_value['NWSTAT1'][:5] == '18158':
        crimtype += ['human trafficking']
    elif row_value['NWSTAT1'][:5] == '18159':
        crimtype += ['human trafficking']
    elif row_value['NWSTAT1'][:6] == '182250':
        crimtype += ['sex offender registry']
    elif row_value['NWSTAT1'][:6] == '182251':
        crimtype += ['sexual abuse of children']
    elif row_value['NWSTAT1'][:5] == '18225':
        crimtype += ['child abuse']
    elif row_value['CRIMETYPE'] == '21':
        crimtype += ['drugs']
    elif row_value['CRIMETYPE'] == '26':
        crimtype += ['taxes']
    elif row_value['CRIMETYPE'] == '42':
        crimtype += ['credit extortion']
    elif row_value['CRIMETYPE'] == '16':
        crimtype += ['conservation']
    elif row_value['CRIMETYPE'] == '29':
        crimtype += ['labor']
    else:
        crimtype += ['other']
tn['CRIMETYPE'] = crimtype
tn
```

```python
violence = []
for row_index, row_value in tn.iterrows():
    if row_value['ADJ_D1'] > 0:
        violence += ['violent']
    elif row_value['NWSTAT1'][:5] == '18111':
        violence += ['violent']
    elif row_value['NWSTAT1'][:5] == '18158':
        violence += ['violence']
    elif row_value['NWSTAT1'][:6] == '182251':
        violence += ['sexual abuse']
    else:
        violence += ['other']
tn['VIOLENCE'] = violence
tn
```
