### Columns to Drop:
- MARRIED missing from all (can be dropped)
- OFFTYPSB missing from all (can be dropped)
- FSASV missing from 2015-2018. Think this info is accounted for in the SAFETY and SAFE cols. Can be dropped.
- SORFORM not missing from any, but needs to be removed, due to not relevant
- WEAPSCHI: indicates if soc enhancement for weapons is present, but it's a version of the weapsoc field. i think we can drop this. 
### Columns to Keep:
- CASETYPE: missing 2015-2017. felony or misdemeanor. Leave this in, may be able to calculate this later.
- DTGDL: missing 2015-2017. identifies cases with one of the "Big 7" drug guidelines as the primary guideline high. Will leave in.
- OFFGUIDE: missing 2015-2017. "Primary type of crime for the case
- SENSPCAP: missing 2015-2017. total prison sentence in months plus alternatives with zeros. i can't tell if this is the actual sentence or maximum possible sentence. keeping.
- SENTRNGE: missing 2015-2017. this is the 2018-2019 col that replaces the bookercd column.
- SENTTCAP: missing 2015-2017. total sentence. this replaced SENTTOT0.
- MNTHDEPT: missing 2015-2017. statistical measure. important, keeping.
- PCNTDEPT: missing 2015-2017. statistical measure. def worth keeping, may be able to manually calculate.
- BOOKERCD: missing from 2018-2019. Field was replaced by SENTRNGE for these years. Values have different meanings, so need to keep both of these columns.
- SENTTOT0: missing from 2018-2019. this was replaced by SENTTCAP.
